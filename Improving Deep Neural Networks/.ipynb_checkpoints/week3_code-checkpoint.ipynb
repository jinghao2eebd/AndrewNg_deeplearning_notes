```python
import tensorflow as tf
import numpy as np
```

```python
w = tf.Variable(0., dtype=tf.float32)
cost = w**2 - 10*w +25
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
sess.run(w)
```

```python
sess.run(train)
print(sess.run(w))
```

```python
for i in range(1000):
    sess.run(train)
sess.run(w)
```

```python
w = tf.Variable(0., dtype=tf.float32)
x = tf.placeholder(tf.float32, [3, 1])
cost = x[0][0] * w**2 + x[1][0]*w +x[2][0]
coefficient = np.array([[1.], [-10.], [25.]])
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
sess.run(w)
```

```python
sess.run(train, feed_dict={x: coefficient})
sess.run(w)
```

```python
for i in range(1000):
    sess.run(train, feed_dict={x: coefficient})
sess.run(w)
```
